```python
import import_ipynb

import cores_pessoais
import argparse
import os

def main():
    # Crie uma instância para receber argumentos.
    parser = argparse.ArgumentParser(description = 'Please input your image.')

    # Registrar valores de argumentos de entrada
    parser.add_argument('--image', required = False, help='input .jpg or .png file')
    parser.add_argument('--dir', required = False, help='input image directory')

    # Armazene os valores dos argumentos de entrada em args.
    args = parser.parse_args(['--image', '../static/sua_foto.jpg'])

    ##################################
    #         Imagem unica         #
    ##################################
    if args.image != None:
        imgpath = args.image
        cores_pessoais.analysis(imgpath)

    ##################################
    #  multiplas imagens de um diretorio  #
    ##################################
    elif args.dir != None:
        dirpath = args.dir
        imgs = os.listdir(dirpath)
        for imgpath in imgs:
            #print(os.path.join(dirpath, imgpath))
            cores_pessoais.analysis(os.path.join(dirpath, imgpath))

if __name__ == '__main__':
    main()
```
